```python
#Renamed columns for more professional and grammatically correct names

df.rename(columns={
    'movieID' : 'Movie ID',
    'title' : 'Title',
    'genres' : 'Genres'
}, inplace=True)

df.columns
```

```python
#Split genres string into lists using "|" separator
df['Genres_list'] = df['Genres'].apply(lambda x : x.split('|'))

df['Genres_list']
```

```python
#Basic Genre Frequency Analysis 


genre_counts = df.explode('Genre_list')['Genre_list'].value_counts()
print(genre_counts)
```
